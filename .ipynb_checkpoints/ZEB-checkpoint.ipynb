```python
import pandas as pd
import numpy as np
import datetime
import pytz
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
import seaborn as sns
import statsmodels.api as sm

print(pd.__version__)
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
discfile = 'data/ZEB Energy Data for SinberBEST.xlsx'
weatherfile = 'data/ZEB_exported_SGP_PAYALEBAR.xlsx'
exportfile = 'ZEB_exported'
```

```python
# 读取能耗数据
read_sheetname_BCA = 'ALL DATA'
read_parse_cols = 'B,C,D'
skiprows = 6
```

```python
read_energy = pd.read_excel(discfile, parse_cols=read_parse_cols, sheetname=read_sheetname_BCA, skiprows=skiprows,
                            names=['date', 'PVGen', 'Consumption'])
read_energy.index = pd.to_datetime(read_energy.iloc[:, 0], dayfirst=True)
read_energy = read_energy._get_numeric_data()  # 移除date列
read_energy = read_energy.replace(0, np.nan)  # 处理Nan
read_energy[read_energy < 10] = np.nan  # 处理异常值(A trick here)
# 获取含有无效值的行
tmp_index = read_energy.isnull().any(axis=1)
# 通过插值替代NaN
read_energy_inter = read_energy.interpolate(method='index')
# 计算每日的能源
read_energy_perday = read_energy_inter.diff()
read_energy_perday = read_energy_perday.drop(read_energy_perday.index[tmp_index])
read_energy_perday
```

```python
# 读取天气数据
read_sheetname_station = 'SGP_PAYALEBAR'
read_parse_cols = "B,D,F,N,O,P,Q"
# read_index_col = None
read_weather = pd.read_excel(discfile, parse_cols=read_parse_cols, sheetname=read_sheetname_station)
read_weather.index = pd.to_datetime(read_weather['valid'], yearfirst=True, format='%Y%m%d')
# 替换云量数值
# vals_to_replace = {'M': 0.0, 'CLR': 0.0, 'FEW': 25.0, 'SCT': 50.0, 'BKN': 75.0, 'OVC': 100.0}
read_weather
```

```python
vals_to_replace = {'M': 0.0, 'CLR': 0.0, 'FEW': 1.0, 'SCT': 2.0, 'BKN': 3.0, 'OVC': 4.0, 'VV': 0.0, '': 0.0}
read_weather = read_weather.replace(vals_to_replace)
read_weather.iloc[:, 3:7] = read_weather.iloc[:, 3:7].apply(pd.to_numeric, errors="coerce")  # 转换数据
read_weather = read_weather.drop('valid', 1)
```

```python
# 设置时区，转换时间
read_weather.index.tz = 'UTC'
read_weather = read_weather.tz_convert('Asia/Singapore')
# 全年来看，新加坡日出日落约在7:00-19:00，筛选出这段时间
read_weather = read_weather.ix[read_weather.index.indexer_between_time(datetime.time(7), datetime.time(19))]

chk_weather = read_weather.groupby(read_weather.index.date).mean()
chk_weather.index = pd.to_datetime(chk_weather.index)
chk_weather
```

```python
# 合并数据
result = pd.concat([chk_weather, read_energy_perday], axis=1).dropna()
result.columns =[x.strip() for x in result.columns]
result
```

```python
result['PVGen_cat'] = pd.qcut(result["PVGen"], 5, labels=False)
result
```

```python
sns.pairplot(result, hue='PVGen')
sns.plt.show()
```

```python
sns.lmplot(x="PVGen", y="Consumption", data=result)
sns.plt.show()
```

```python
sns.lmplot(y="PVGen", x="Celcius", data=result)
sns.plt.show()
```

```python
sns.lmplot(x="relh", y="Celcius", data=result)
sns.plt.show()
```

```python
X = result[["Consumption","Celcius","relh"]]
X = sm.add_constant(X) # another way to add a constant row for an intercept
y = result["PVGen"]

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
regression_x = result.iloc[:,0:6].values
regression_x
```

```python
regression_y = result.ix[:,6].values
regression_y
```

```python
x_train = regression_x[:-20]
x_test = regression_x[-20:]
y_train = regression_y[:-20]
y_test = regression_y[-20:]
regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % np.mean((regr.predict(x_test) - y_test) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x_test, y_test))
```
